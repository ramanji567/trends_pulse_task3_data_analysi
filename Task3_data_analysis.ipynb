```python
score = df["score"]
```

```python
import numpy as np
amount = np.array(score)
#a = amount.mean()
b = amount.std()
print(b)
```

```python
max_score = amount.max()
max_score
```

```python
min_score = amount.min()
min_score
```

```python
most_stories =df['category'].value_counts()['technology']
```

```python
df["num_comments"].max()
```

```python
most_commented_story = df.loc[df["num_comments"].idxmax()]
most_title = most_commented_story["title"]
no_of_comments =most_commented_story["num_comments"]
```
